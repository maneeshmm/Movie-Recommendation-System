```python
final_dataset = final_dataset.loc[:, no_movies_voted[no_movies_voted > 50].index] #Keeps only users who have rated more than 50 movies.
```

```python
final_dataset
```

```python
from scipy.sparse import csr_matrix
csr_data = csr_matrix(final_dataset.values) #did filter out movies with fewer than 10 ratings, which removes the least-rated movies. However, even among the remaining "popular" movies, not every active user has rated every movie.
final_dataset.reset_index(inplace = True) # the final_dataset has a lot of 0s (unrated movies), so converting it into a sparse matrix using csr_matrix makes processing more efficient
```

```python
print(csr_data) #When you use csr_matrix (Compressed Sparse Row matrix) from scipy.sparse, it only stores the nonzero values and their positions, making it more efficient.
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute', n_neighbors = 20, n_jobs = -1)
```

metric='cosine' → Uses cosine similarity to measure how similar two items are.
algorithm='brute' → Uses a brute-force approach (compares every point to every other).
n_neighbors=20 → Finds the 20 most similar items (movies or users).
n_jobs=-1 → Uses all available CPU cores for faster computation.

```python
knn.fit(csr_data)
```

```python
def get_recommendation(movie_name):
    movie_list = movies[movies['title'].str.contains(movie_name, case=False, na=False)] 
    if movie_list.empty:
        return "Movie not found..."

    movie_id = movie_list.iloc[0]['movieId']
    idx = final_dataset.index[final_dataset['movieId'] == movie_id].tolist()
    
    if not idx:
        return "Movie not found in dataset..."
    
    distances, indices = knn.kneighbors(csr_data[idx[0]], n_neighbors=11)
    return movies.loc[movies['movieId'].isin(final_dataset.iloc[indices[0][1:]]['movieId']), 'title'].tolist()
```

movies['title'].str.contains(movie_name, case=False, na=False)
This checks if any movie title in the movies dataset contains movie_name.
case=False → Ignores case (e.g., "avatar" = "Avatar").
na=False → Prevents errors from missing values (NaN).
movies[...] → Returns a filtered DataFrame (movie_list) containing matching movie(s).

if movie_list.empty:
    return "Movie not found..."
If no movie matches the search, return "Movie not found..."

movie_id = movie_list.iloc[0]['movieId']
.iloc[0] → Selects the first matching movie (to keep it simple).
['movieId'] → Extracts the Movie ID from the row.

idx = final_dataset.index[final_dataset['movieId'] == movie_id].tolist()
Looks for movieId in final_dataset and gets its index.


if not idx:
    return "Movie not found in dataset..."
If the movie isn’t found in final_dataset, return "Movie not found in dataset...".


distances, indices = knn.kneighbors(csr_data[idx[0]], n_neighbors=11)   Find Similar Movies Using KNN
inds the 10 most similar movies (excluding the input movie).
indices → Indices of similar movies.
distances → Cosine similarity distances.


return movies.loc[movies['movieId'].isin(final_dataset.iloc[indices[0][1:]]['movieId']), 'title'].tolist()
Gets the movie titles of the recommended movies using their movieId.

```python
import gradio as gr

def recommend_movies(movie_name):
    df = get_recommendation(movie_name)
    if isinstance(df, pd.DataFrame):
        return df.to_string(index=False)
    else:
        return df
app = gr.Interface(
    fn = recommend_movies, 
    inputs = "text",
    outputs = "text", 
    title = "Movie Recommendation System",
    description = "Enter a movie name to get a list of recommended movies"
)

app.launch()
```
